```python
def hitung_angsuran(harga, tenor, bunga=0.03):
    total_bunga = harga * bunga * tenor
    total_bayar = harga + total_bunga
    angsuran_bulanan = total_bayar / tenor

    return {
        "angsuran_per_bulan": round(angsuran_bulanan, 2),
        "total_bunga": round(total_bunga, 2),
        "total_pembayaran": round(total_bayar, 2)
    }
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from datetime import datetime
import pandas as pd
import re

def parse_harga(input_harga):
    teks = input_harga.lower().strip().replace(" ", "")

    if "juta" in teks or "jt" in teks:
        angka = re.findall(r"\d+(\.\d+)?", teks)
        if not angka:
            raise ValueError("Format harga tidak valid")
        return float(angka[0].replace(".", "")) * 1_000_000

    teks = teks.replace(".", "").replace(",", "")
    if teks.isdigit():
        return float(teks)

    raise ValueError("Format harga tidak dikenali")

def hitung_angsuran(harga, tenor, bunga_bulanan=0.055):
    total_bunga = harga * bunga_bulanan * tenor
    total_bayar = harga + total_bunga
    angsuran = total_bayar / tenor
    return angsuran, total_bunga, total_bayar

nama_customer = widgets.Text(description="Customer:")
produk = widgets.Text(description="Produk:")
harga = widgets.Text(
    description="Harga (Rp):",
    placeholder="contoh: 16 jt / 16.000.000"
)
tenor = widgets.Dropdown(
    options=[6, 12, 18, 24],
    description="Tenor:"
)

button = widgets.Button(
    description="Hitung Angsuran",
    button_style="success"
)

output = widgets.Output()

def on_button_clicked(b):
    with output:
        clear_output()
        try:
            harga_value = parse_harga(harga.value)
            angsuran, total_bunga, total_bayar = hitung_angsuran(
                harga_value, tenor.value
            )

            print("=== HASIL SIMULASI ANGSURAN ===")
            print(f"Customer        : {nama_customer.value}")
            print(f"Produk          : {produk.value}")
            print(f"Harga Produk    : Rp {harga_value:,.0f}")
            print(f"Tenor           : {tenor.value} bulan")
            print(f"Angsuran/Bulan  : Rp {angsuran:,.0f}")
            print(f"Total Bunga     : Rp {total_bunga:,.0f}")
            print(f"Total Bayar     : Rp {total_bayar:,.0f}")

        except Exception as e:
            print("‚ùå Error:", e)

button.on_click(on_button_clicked)

display(
    widgets.VBox([
        nama_customer,
        produk,
        harga,
        tenor,
        button,
        output
    ])
)

```

```python
import streamlit as st
import pandas as pd
import re
from datetime import datetime
import os

st.set_page_config(
    page_title="Kalkulator Angsuran",
    layout="centered"
)

DATA_FILE = "data_customer_angsuran.xlsx"

def parse_harga(input_harga):
    teks = input_harga.lower().strip().replace(" ", "")

    if "juta" in teks or "jt" in teks:
        angka = re.findall(r"\d+(\.\d+)?", teks)
        return float(angka[0].replace(".", "")) * 1_000_000

    teks = teks.replace(".", "").replace(",", "")
    return float(teks)

def hitung_angsuran(harga, tenor, bunga=0.055):
    total_bunga = harga * bunga * tenor
    total_bayar = harga + total_bunga
    angsuran = total_bayar / tenor
    return angsuran, total_bunga, total_bayar

st.title("üì± Kalkulator Angsuran Produk")
st.write("Bunga **5,5% per bulan (flat)**")

nama_customer = st.text_input("Nama Customer")
produk = st.text_input("Nama Produk")
harga_input = st.text_input("Harga Produk (contoh: 16 jt / 16.000.000)")
tenor = st.selectbox("Tenor (bulan)", [6, 12, 18, 24])

if st.button("Hitung Angsuran"):
    try:
        harga = parse_harga(harga_input)
        angsuran, total_bunga, total_bayar = hitung_angsuran(harga, tenor)

        st.success("‚úÖ Perhitungan Berhasil")
        st.write(f"**Harga Produk**: Rp {harga:,.0f}")
        st.write(f"**Angsuran / Bulan**: Rp {angsuran:,.0f}")
        st.write(f"**Total Bunga**: Rp {total_bunga:,.0f}")
        st.write(f"**Total Bayar**: Rp {total_bayar:,.0f}")

        data = {
            "Tanggal": datetime.now().strftime("%Y-%m-%d %H:%M"),
            "Customer": nama_customer,
            "Produk": produk,
            "Harga": harga,
            "Tenor": tenor,
            "Angsuran/Bulan": angsuran,
            "Total Bunga": total_bunga,
            "Total Bayar": total_bayar
        }

        df_new = pd.DataFrame([data])

        if os.path.exists(DATA_FILE):
            df_old = pd.read_excel(DATA_FILE)
            df = pd.concat([df_old, df_new], ignore_index=True)
        else:
            df = df_new

        df.to_excel(DATA_FILE, index=False)

    except Exception as e:
        st.error(f"‚ùå {e}")

st.divider()
st.subheader("üì• Download Data Customer")

if os.path.exists(DATA_FILE):
    df_download = pd.read_excel(DATA_FILE)
    st.download_button(
        label="Download Excel",
        data=df_download.to_excel(index=False),
        file_name="data_customer_angsuran.xlsx",
        mime="application/vnd.openxmlformats-officedocument.spreadsheetml.sheet"
    )
else:
    st.info("Belum ada data tersimpan")
```
